## Atributos e Variável Alvo
Em problemas de aprendizado supervisionado, é comum separar o dataset em dois grupos: as variáveis explicativas (atributos) e a variável alvo (target). As variáveis explicativas são usadas pelo modelo para aprender padrões enquanto a variável alvo é aquilo que desejamos prever.

```python
# Separar atributos e variável alvo.
X = df.drop(columns=["vendas_sorvete"])
y = df["vendas_sorvete"]

X.head(), y.head()
```

## Escalonamento
Em muitos algoritmos de machine learning, as variáveis precisam estar em escalas comparáveis. Quando os atributos possuem magnitudes muito diferentes (por exemplo, idade em dezenas e renda em milhares), muitos modelos podem dar mais peso às variáveis com valores maiores. O escalonamento resolve esse problema transformando os dados para uma escala comum. Nesta prática utilizaremos a padronização (z-score), que transforma os dados para que tenham média igual a 0 e desvio padrão igual a 1.

## Correlação
A correlação é uma medida estatística que indica o grau de associação linear entre duas variáveis. Valores próximos de 1 ou -1 indicam forte relação linear, enquanto valores próximos de 0 indicam pouca ou nenhuma relação linear. No contexto de seleção de atributos, a correlação pode ajudar a identificar quais variáveis estão mais associadas à variável alvo, servindo como um primeiro critério para avaliar a relevância de cada atributo no problema preditivo.

### Pergunta

```python
# Calcular correlação das variáveis com a variável alvo.
df_corr = df.corr(numeric_only=True)
df_corr["vendas_sorvete"].sort_values(ascending=False)
```

## Informação Mútua
A informação mútua mede o grau de dependência estatística entre duas variáveis. Diferentemente da correlação, ela não se limita a relações lineares e pode detectar dependências mais gerais entre os dados. Em tarefas de seleção de atributos, a informação mútua é frequentemente utilizada para identificar variáveis que fornecem maior quantidade de informação sobre a variável alvo, ajudando a determinar quais atributos podem ser mais úteis para um modelo preditivo.

### Pergunta

10. ordem de importância das variáveis mudou em relação à correlação? Ou seja, as variáveis mais correlacionadas continuam sendo aquelas com maior informação mútua?

```python
# Calcular informação mútua entre os atributos e a variável alvo.
from sklearn.feature_selection import mutual_info_regression
import pandas as pd

mi = mutual_info_regression(X, y, random_state=42)
pd.Series(mi, index=X.columns).sort_values(ascending=False)
```
